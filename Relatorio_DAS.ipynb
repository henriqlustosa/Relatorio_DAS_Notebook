```python
import pandas as pd
import os

# Define the folder path where your Excel files are located
folder_path ='G:\\OneDrive - Default Directory\\OneDrive - rede.sp\\Relatorio_Excel_DAS\\Relatorio_2024\\'

# Get a list of all Excel files in the folder
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

# Initialize an empty list to store DataFrames
df_list = []

# Loop through the Excel files and read them into DataFrames
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)  # Read the Excel file
    df_list.append(df)  # Append the DataFrame to the list

# Concatenate all DataFrames into one
df_combined = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new Excel file
output_file_path = os.path.join(folder_path, 'Relatorio_2023.xlsx')
df_combined.to_excel(output_file_path, index=False)

print(f"Combined Excel file saved as: {output_file_path}")
```
